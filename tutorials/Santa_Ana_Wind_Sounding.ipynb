```python
'''
In this example we will plot a sounding from a custom time the user wishes to analyze. This can be useful for meteorologists who are conducting a 
case study or post-analysis study. In this example we will look at the start of a Santa Ana Wind event in San Diego, CA (NKX). On October 29th, 2023 at 00z onshore winds 
were observed. Notice the winds on the 29th/00z are northwesterly (onshore flow since onshore winds in California are winds with westerly components) in the lowest 5000ft AGL. The RH vs. Height
profile on the 29th/00z shows the marine layer between 2000-3000ft AGL as the blue line sharply drops off between those two elevation points. 

Santa Ana Winds began sometime between the 29th/00z and 30th/00z as the NKX Sounding from October 30th, 2023 at 00z shows weak northeasterly winds in the lowest 5000ft AGL 
(offshore winds which are "dry winds" are winds with easterly components) northeasterly winds are referred to as "Santa Ana Winds." Now look at the RH vs. Height comparison, 
the green line (30th/00z) is very dry in the lowest 5000ft AGL with single digit RH in the lowest 2000-3000ft and low 10s RH from 3000-5000ft AGL, a very dry profile typical
during Santa Ana Winds. Also the hotter temperatures and very dry airmass allow for lapse rates to be near dry-adiabatic (10 degrees Celsius per kilometer) and thus allow for ample
vertical mixing in the lowest 8000ft. One good thing about Santa Ana Wind events due to this is they bring very good air quality since they provide excellent dispersion. 

In Python, the user will call the plot_observed_sounding_custom_date_time from the firewxpy.soundings module and pass in the station_id as a string followed by the date/time of the sounding.

Here is the format:
plot_observed_sounding_custom_date_time(station_id, year, month, day, hour) 

Now let's make our graphic!
'''

plot_observed_sounding_custom_date_time('nkx', 2023, 10, 30, 0) 
```
